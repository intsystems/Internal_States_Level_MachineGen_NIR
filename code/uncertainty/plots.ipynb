```python
import pandas as pd

human_df = pd.read_csv('data/human.csv', index_col=[0])
machine_df = pd.read_csv('data/machine.csv', index_col=[0])

human_df.head()
```

```python
machine_df.msp.describe()
```

```python
human_msp_mean = human_df.msp.mean()
machine_msp_mean = machine_df.msp.mean()

human_msp_median = human_df.msp.median()
machine_msp_median = machine_df.msp.median()

import matplotlib.pyplot as plt

plt.hist(human_df.msp, bins=10, alpha=0.8, color='skyblue', label='Human')
plt.axvline(human_msp_mean, color='skyblue', label='Human mean')
plt.axvline(human_msp_median, color='skyblue', linestyle='--', label='Human median')
plt.hist(machine_df.msp, bins=10, color='indianred', alpha=0.8, label='Machine')
plt.axvline(machine_msp_mean, color='indianred', label='Machine mean')
plt.axvline(machine_msp_median, color='indianred', linestyle='--', label='Machine median')
plt.title('MSP')
plt.legend()
plt.savefig('msp.png')
plt.show()
```

```python
human_msp_perp = human_df.perplexity.mean()
machine_msp_perp = machine_df.perplexity.mean()
plt.hist(human_df.perplexity, bins=10, color='skyblue', alpha=0.8, label='Human')
plt.axvline(human_msp_perp, color='skyblue', label='Human mean')
plt.hist(machine_df.perplexity, bins=10, color='indianred', alpha=0.8, label='Machine')
plt.axvline(machine_msp_perp, color='indianred', label='Machine mean')
plt.title('Perplexity')
plt.legend()
plt.savefig('perplexity.png')
plt.show()
```

```python
human_msp_entropy = human_df.entropy.mean()
machine_msp_entropy = machine_df.entropy.mean()

plt.hist(human_df.entropy, bins=10, color='skyblue', alpha=0.8, label='Human')
plt.axvline(human_msp_entropy, color='skyblue', label='Human mean')
plt.hist(machine_df.entropy, bins=10, color='indianred', alpha=0.8, label='Machine')
plt.axvline(machine_msp_entropy, color='indianred', label='Machine mean')
plt.title('Entropy')
plt.legend()
plt.savefig('entropy.png')
plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(12, 6))
axs[0].hist(human_df.msp, bins=10, alpha=0.8, color='skyblue', label='Human')
axs[0].axvline(human_msp_mean, color='skyblue', label='Human mean')
axs[0].axvline(human_msp_median, color='skyblue', linestyle='--', label='Human median')
axs[0].hist(machine_df.msp, bins=10, color='indianred', alpha=0.8, label='Machine')
axs[0].axvline(machine_msp_mean, color='indianred', label='Machine mean')
axs[0].axvline(machine_msp_median, color='indianred', linestyle='--', label='Machine median')
axs[0].set_ylim(0, 100)

axs[0].set_title('MSP')
axs[1].hist(human_df.perplexity, bins=10, color='skyblue', alpha=0.8, label='Human')
axs[1].axvline(human_msp_perp, color='skyblue', label='Human mean')
axs[1].hist(machine_df.perplexity, bins=10, color='indianred', alpha=0.8, label='Machine')
axs[1].axvline(machine_msp_perp, color='indianred', label='Machine mean')
axs[1].set_title('Perplexity')
axs[1].set_ylim(0, 100)


axs[2].hist(human_df.entropy, bins=10, color='skyblue', alpha=0.8, label='Human')
axs[2].axvline(human_msp_entropy, color='skyblue', label='Human mean')
axs[2].hist(machine_df.entropy, bins=10, color='indianred', alpha=0.8, label='Machine')
axs[2].axvline(machine_msp_entropy, color='indianred', label='Machine mean')
axs[2].set_title('Entropy')
# fig.suptitle('Human and Machine Metrics')
plt.tight_layout()
plt.savefig('metrics.png')
plt.show()
```
